```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df = pd.read_csv("merged_output.csv")
```

```python
data = df.select_dtypes(include=[np.number]).copy()

print("Dataset Shape:", data.shape)
```

```python
data["Total_Power"] = data.sum(axis=1)

if "Alpha_abs" in data.columns and "Beta_abs" in data.columns:
    data["Alpha_Beta_Ratio"] = data["Alpha_abs"] / (data["Beta_abs"] + 1e-6)

if "Theta_abs" in data.columns and "Gamma_abs" in data.columns:
    data["Theta_Gamma_Ratio"] = data["Theta_abs"] / (data["Gamma_abs"] + 1e-6)
```

```python
data = data.rolling(window=2, min_periods=1).mean()
```

```python
original_data = data.copy()

np.random.seed(42)
mask = np.random.rand(*data.shape) < 0.05

missing_data = data.copy()
missing_data[mask] = np.nan

print("Missing values introduced:", np.sum(mask))
```

```python
scaler = StandardScaler()
scaled_missing = scaler.fit_transform(missing_data)
```

```python
def evaluate_imputation(original, imputed, mask):
    original_vals = original.values[mask]
    imputed_vals = imputed.values[mask]

    rmse = np.sqrt(mean_squared_error(original_vals, imputed_vals))
    mae = mean_absolute_error(original_vals, imputed_vals)
    r2 = r2_score(original_vals, imputed_vals)

    accuracy = max(0, r2 * 100)

    return rmse, mae, r2, accuracy
```

```python
print("\nRunning Random Forest MICE...")

rf_estimator = RandomForestRegressor(
    n_estimators=50,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)

rf_mice = IterativeImputer(
    estimator=rf_estimator,
    max_iter=10,
    random_state=42
)

rf_scaled = rf_mice.fit_transform(scaled_missing)
rf_imputed = scaler.inverse_transform(rf_scaled)

rf_df = pd.DataFrame(rf_imputed, columns=data.columns)

rf_metrics = evaluate_imputation(
    original_data,
    rf_df,
    mask
)
```

```python
results = pd.DataFrame({
    "Method": [
        "Random Forest MICE"
    ],
    "RMSE": [
        rf_metrics[0]
    ],
    "MAE": [
        rf_metrics[1]
    ],
    "R2 Score": [
        rf_metrics[2]
    ],
    "Accuracy (%)": [
        rf_metrics[3]
    ]
})

print("\n========== FINAL RESULTS ==========")
print(results)

best_dataset = rf_df

best_dataset.to_csv("best_imputed_dataset.csv", index=False)
results.to_csv("imputation_results.csv", index=False)
```
